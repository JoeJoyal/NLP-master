```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
```

```python
import nltk
```

```python
nltk.download('punkt_tab')
```

```python
sentence = [
    "I love deep learning",
    "Deep learning is part of AI",
    "Word embedding capture meaning",
    "I love NLP"
]
```

```python
word_tokenize = [word_tokenize(i.lower()) for i in sentence]
```

```python
list(word_tokenize)
```

```python
model_cbow = Word2Vec(sentences=word_tokenize, vector_size=100, window=2, min_count=1, sg=0)
```

```python
model_skip_gram = Word2Vec(sentences=word_tokenize, vector_size=100, window=2, min_count=1, sg=1)
```

```python
model_cbow.wv.most_similar('love')
```

```python
model_skip_gram.wv.most_similar('love')
```
