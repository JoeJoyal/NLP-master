```python
import re
import joblib
import string
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix
```

```python
df =pd.read_csv('multiclass-text-data.csv')
```

```python
df.head()
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub(r'\s+',' ', text).strip()
    return text
```

```python
df['cleaned_text'] = df['text'].apply(clean_text)
```

```python
df
```

```python
vectorizer = TfidfVectorizer()
x = vectorizer.fit_transform(df['cleaned_text'])
```

```python
x.toarray()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit_transform(df['cleaned_text'])
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['category_encoded'] = le.fit_transform(df['cleaned_text'])
```

```python
df
```

```python
df.tail(20)
```

```python
y = df['category_encoded']
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model = MultinomialNB()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
joblib.dump(model,'MultinomialNB_spam.pkl')
```

```python
joblib.dump(vectorizer,'tfidf_MultinomialNB_spam.pkl')
```
